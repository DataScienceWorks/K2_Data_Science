```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1.

```python
# Generate 3 arrays x, y1, y2 of 500 values
num_count = 500

x = np.linspace(-2*np.pi, 2*np.pi, num=num_count)
y1 = np.sin(x)
y2 = np.cos(x)
```

## 2.

```python
# Plot y1, y2 over x on the same graph
plt.plot(x, y1, x, y2)
plt.show()
```

## 3.

```python
# Plot y1, y2 over x on the same graph with tighter x and y limits
plt.plot(x, y1, x, y2)
plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)
plt.show()
```

## 4.

```python
# Plot y1, y2 over x on the same graph with tighter x and y limits and change color and line settings
scale = 1.25

plt.plot(x, y1, 'r--', x, y2, 'b:')
plt.xlim(scale*x.min(), scale*x.max())
plt.ylim(scale*y1.min(), scale*y1.max())
plt.show()
```

## 5.

```python
# Plot y1, y2 over x on the same graph with ticks and labels for important points
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.plot(x, np.zeros(len(x)))

# Create array of points for roots of sin
roots = [-np.pi, 0, np.pi]
sin_roots = np.sin(roots)

# Plot sin_roots
plt.plot(roots, sin_roots, 'o')

# Label sin_roots
plt.annotate("Sin root", xy=(roots[0], sin_roots[0]), xytext=(roots[0], .75), color='red', weight='heavy', arrowprops={'facecolor': 'red'})
plt.annotate("Sin root", xy=(roots[1], sin_roots[1]), xytext=(roots[1], .75), color='red', weight='heavy', arrowprops={'facecolor': 'red'})
plt.annotate("Sin root", xy=(roots[2], sin_roots[2]), xytext=(roots[2], .45), color='red', weight='heavy', arrowprops={'facecolor': 'red'})
plt.xlim(-5.0, 5.0)
plt.ylim(-1.0, 1.0)
plt.legend()

plt.show()
```

## 6.

```python
# Plot y1, y2 over x on the same graph with ticks and labels for important points
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.plot(x, np.zeros(len(x)))

# Create array of points for roots of sin
roots = [-np.pi, 0, np.pi]
sin_roots = np.sin(roots)

# Plot sin_roots
plt.plot(roots, sin_roots, 'o')

# Label sin_roots
plt.xlim(-5.0, 5.0)
plt.ylim(-1.0, 1.0)
plt.legend(frameon=False)

# Change the spines from box to cross

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
# Question 1
air_crash_df = pd.read_csv('airAccs.csv')
air_crash_df.head()
```

```python
air_crash_df.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
air_crash_df.head()
```

```python
# Question 2
air_crash_df.plot(kind='hist', y='Dead', bins=50)
plt.axis([0, 165, 0, 3300])
plt.xticks(np.arange(0, 180, 20))
plt.title('Deaths in an Airline Accident')
plt.xlabel('Deaths')
plt.ylabel('Count')

plt.show()
```

```python
# Question 3
ts = air_crash_df.copy()
ts = ts.set_index(pd.DatetimeIndex(ts.Date)).drop('Date', axis=1)
t = pd.date_range(start=air_crash_df.Date[1], end=air_crash_df.Date[len(air_crash_df) - 1])

ts = ts.loc[t]
ts = ts.fillna(0)
ts['year'] = ts.index.year
ts_yearly_tot = ts.groupby('year').sum().Dead
```

```python
plt.plot(ts_yearly_tot, linewidth=2.5)
plt.xlim(ts_yearly_tot.index[0], ts_yearly_tot.index[len(ts_yearly_tot.index) - 1])
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.title('Total Deaths per Year')

plt.show()
```

```python
# Question 4
passenger_df = pd.read_csv('pass_data.csv', encoding='ISO-8859-1')
passenger_df.head()
```

```python
passenger_df.dtypes
```

```python
data_pass = passenger_df.iloc[:, 4:].sum()
data_pass = pd.DataFrame({'year': data_pass.index.astype(float),
                         'Passengers': data_pass})
data_acc_temp = air_crash_df[['Date', 'Dead']]
data_acc_temp = data_acc_temp.set_index(pd.DatetimeIndex(air_crash_df.Date))
data_acc_temp['year'] = data_acc_temp.index.year
data_acc_temp = data_acc_temp.groupby('year').sum()
data_acc_temp = pd.DataFrame({'year': data_acc_temp.index.astype(float),
                             'Dead': data_acc_temp.Dead})

data_all = pd.merge(data_pass, data_acc_temp)[['year', 'Passengers', 'Dead']]

data_all = data_all[(data_all.Passengers != 0) & (data_all.Dead != 0)]
data_all.head()
```

```python
# Question 5
data_all['per_M'] = (data_all.Dead / data_all.Passengers) * 1000000
data_all['year'] = [int(y) for y in data_all.year]

plt.figure(1)
plt.subplot(3, 1, 1)
plt.plot(data_all.year, data_all.Passengers/ 1000000)
plt.ylabel('Millions of Passengers')

plt.figure(2)
plt.subplot(3, 1, 2)
plt.plot(data_all.year, data_all.Dead)
plt.ylabel('Deaths')

plt.figure(3)
plt.subplot(3, 1, 3)
plt.plot(data_all.year, data_all.per_M)
plt.ylabel('Deaths per Million')

plt.show()
```

```python
plt.figure(1)
plt.subplot(3, 1, 1)
plt.bar(data_all.year, data_all.Passengers/ 1000000)
plt.ylabel('Millions of Passengers')

plt.figure(2)
plt.subplot(3, 1, 2)
plt.bar(data_all.year, data_all.Dead)
plt.ylabel('Deaths')

plt.figure(3)
plt.subplot(3, 1, 3)
plt.bar(data_all.year, data_all.per_M)
plt.ylabel('Deaths per Million')

plt.show()
```

```python
# Question 6
data_all_d = data_all
data_all_d['dec'] = data_all_d.year // 10 * 10
data_all_d = data_all_d.groupby('dec').sum()[['Dead']]

labels = '1970', '1980', '1990', '2000', '2010'
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
plt.pie(data_all_d.Dead, colors=colors, autopct = '%1.1f%%', shadow=True, startangle=140)
plt.title('Fraction of Airline Deaths by Decade')
plt.axis('equal')
plt.legend(labels, loc='best')
plt.show()
```
